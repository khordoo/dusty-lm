# DustyLM: Chat with Dusty

Download a pre-trained 8M parameter DustyLM checkpoint and chat with it in under 30 seconds. Run all cells

**Model:** [mkhordoo/dusty-8m-sft](https://huggingface.co/mkhordoo/dusty-8m-sft)  
**No GPU required.** Runs on CPU, Colab, or any machine with Python.




### Setup 
* **If on Google Colab:** Simply run the cell below.

* **If on locall machine**:
    1. run `uv sync` in repository root, then select the `.venv` as your kernel. 
    2. run the cell below

```python
import sys

if "google.colab" in sys.modules:
    !pip install -q "git+https://github.com/khordoo/dusty-lm.git" torch tokenizers huggingface_hub
else:
    print("💻 Local environment detected.")
    print(
        "Make sure you have run `uv sync` in the terminal and selected the `.venv` kernel."
    )

from huggingface_hub import hf_hub_download

checkpoint_path = hf_hub_download(
    repo_id="mkhordoo/dusty-8m-sft",
    filename="model.pt",
)
tokenizer_path = hf_hub_download(
    repo_id="mkhordoo/dusty-8m-sft",
    filename="tokenizer.json",
)

print("✔ Model downloaded successfully.")
```

```python
import torch
from dustylm.inference import Inference

engine = Inference(
    checkpoint_path=checkpoint_path,
    tokenizer_path=tokenizer_path,
    device="cuda" if torch.cuda.is_available() else "cpu",
)


def chat(prompt):
    return engine.chat_completion(
        messages=[{"role": "user", "content": prompt}],
        max_tokens=64,
        temperature=0.6,
        top_p=0.8,
    )["choices"][0]["message"]["content"].strip()


# Quick test
for prompt in ["hi dusty", "are you scared?", "what do you dream about?"]:
    print(f"You> {prompt}")
    print(f"Dusty> {chat(prompt)}\n")
```

```python
# Interactive chat: type your messages
#
# Note: VS Code's notebook interface does not support ``input()``.
# Use the static demo cell above instead. This cell works on Colab
# and in the terminal (uv run python -m dustylm.inference).
while True:
    try:
        prompt = input("You> ").strip()
    except (KeyboardInterrupt, EOFError):
        break
    if not prompt or prompt.lower() in {"quit", "exit", "q"}:
        break
    print(f"Dusty> {chat(prompt)}\n")
```
